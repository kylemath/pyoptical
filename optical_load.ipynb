```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

# Load in example data file and show top rows and some columns

```python
filename = "data/trigs001a.001"
data = pd.read_csv(filename, skiprows=188, sep='\t')

data.tail()
```

# Practice plotting a few channels


```python
data[["A-Ph1", "A-Ph2", "A-Ph3", "A-Ph4", "A-Ph5"]].plot()
plt.show()
```

# Practice looping through channels and plotting each

```python
detectors = ['A', 'B', 'C', 'D']
fig, ax = plt.subplots()

for detector in detectors:
    for x in np.arange(32):
        thischan = detector + '-DC' + str(x+1)
        tempchan = (data[thischan] - data[thischan].mean()) / data[thischan].std()
        tempchan.plot(ax=ax)

plt.show()
```
